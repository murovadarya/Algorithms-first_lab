```python
import numpy as np
import time
import matplotlib.pyplot as mp
```

```python
time_of_mult = []
matrx_volume = []

for i in range(1, 201):
    X = np.random.rand(i, i)
    Y = np.random.rand(i, i)

    result = np.empty((i, i))


    start_time = time.time()
    for i in range(len(X)):
        # iterate through columns of Y
        for j in range(len(Y[0])):
            # iterate through rows of Y
            for k in range(len(Y)):
                result[i][j] += X[i][k] * Y[k][j]
    finish_time = time.time()
    result_time = finish_time - start_time
    matrx_volume.append(i)
    time_of_mult.append(result_time)

print(time_of_mult)

```

```python
model3 = np.poly1d(np.polyfit(matrx_volume, time_of_mult, 3))
```

```python
matrx_volume = np.array(matrx_volume)
coeffs = np.polyfit(matrx_volume, time_of_mult, 3)
fit = np.poly1d(coeffs)
# equation = np.poly1d(fit)
# print(equation)

fit_curve = coeffs[0] * matrx_volume**3 - coeffs[1] * matrx_volume**2 + coeffs[2] * matrx_volume + coeffs[3]
# print(fit_curve)
```

```python
mp.plot(matrx_volume, time_of_mult, label='Время выполнения умножения')
mp.plot(fit_curve, label='Теоретическая прямая')
mp.xlabel("Размер матрицы")
mp.ylabel("Время")
mp.legend()
mp.show()
```
